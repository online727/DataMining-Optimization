```python
import os
import re
import jieba
import fitz
import logging

jieba.setLogLevel(logging.ERROR)
```

```python
with open("stopword.txt", "r", encoding="utf-8") as f:
    stopwords = list(set(f.read().splitlines()))
stopwords[:5]
```

```python
os.chdir("../Data")
reports_path = "AnnualReports"
all_reports = os.listdir(reports_path)
all_reports[:5]
```

```python
def get_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    doc_text = ""
    for page_num in range(doc.page_count):
        page = doc.load_page(page_num)
        doc_text += page.get_text("text")
    clean_text = re.sub(r'[^\u4e00-\u9fa5]', '', doc_text)
    split_words = jieba.lcut(clean_text)
    filtered_words = [word for word in split_words if word not in stopwords]
    del doc, doc_text, clean_text, split_words
    return filtered_words

filtered_words = get_text_from_pdf(os.path.join(reports_path, all_reports[0]))
filtered_words[:5]
```

```python
from tqdm import tqdm

save_path = "FilteredReports"
os.makedirs(save_path, exist_ok=True)

tqdm_bar = tqdm(all_reports, desc="Processing PDFs")
for report in tqdm_bar:
    try:
        filtered_words = get_text_from_pdf(os.path.join(reports_path, report))
        with open(os.path.join(save_path, report.replace(".pdf", ".txt")), "w", encoding="utf-8") as f:
            f.write(" ".join(filtered_words))
        tqdm_bar.set_postfix_str(f"Processed {report}")
    except Exception as e:
        tqdm_bar.write(f"Error processing {report}: {e}")
        continue
```
